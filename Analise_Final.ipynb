## Criando funções de consulta diretamente no banco de dados para o ano de 2020 e 2021
Objetivos:
*   Agrupar os dados por estado
*   Somar a quantidade de exportações de cada produto
*   Trazer apenas os dados necessario para o objetivo da análise




```python
def consulta_exportacao(ano):
  if ano==2020:
    # Consulta agupando os dados por estado e somando quantidade de exportações
    consulta_2020 = """
    SELECT exportacoes.CO_ANO, uf.SG_UF, exportacoes.CO_NCM, ncm.NO_NCM_POR,
          SUM(exportacoes.QT_ESTAT) AS quantidade_total_exportada
    FROM exportacoes
    INNER JOIN ncm ON exportacoes.CO_NCM = ncm.CO_NCM
    INNER JOIN uf ON exportacoes.SG_UF_NCM = uf.SG_UF
    WHERE exportacoes.CO_ANO = 2020
    GROUP BY uf.SG_UF, exportacoes.CO_NCM
    ORDER BY uf.SG_UF, quantidade_total_exportada DESC;
    """
    return consulta_2020
  elif ano==2021:
    # Consulta agupando os dados por estado e somando quantidade de exportações
    consulta_2021 = """
    SELECT exportacoes.CO_ANO, uf.SG_UF, exportacoes.CO_NCM, ncm.NO_NCM_POR,
        SUM(exportacoes.QT_ESTAT) AS quantidade_total_exportada
    FROM exportacoes
    INNER JOIN ncm ON exportacoes.CO_NCM = ncm.CO_NCM
    INNER JOIN uf ON exportacoes.SG_UF_NCM = uf.SG_UF
    WHERE exportacoes.CO_ANO = 2021
    GROUP BY uf.SG_UF, exportacoes.CO_NCM
    ORDER BY uf.SG_UF, quantidade_total_exportada DESC;
    """
    return consulta_2021

def consulta_importacoes(ano):
  if ano==2020:
    # Consulta agupando os dados por estado e somando quantidade de exportações
    consulta_2020 = """
    SELECT importacoes.CO_ANO, uf.SG_UF, importacoes.CO_NCM, ncm.NO_NCM_POR,
          SUM(importacoes.QT_ESTAT) AS quantidade_total_importada
    FROM importacoes
    INNER JOIN ncm ON importacoes.CO_NCM = ncm.CO_NCM
    INNER JOIN uf ON importacoes.SG_UF_NCM = uf.SG_UF
    WHERE importacoes.CO_ANO = 2020
    GROUP BY uf.SG_UF, importacoes.CO_NCM
    ORDER BY uf.SG_UF, quantidade_total_importada DESC;
    """
    return consulta_2020
  elif ano==2021:
    # Consulta agupando os dados por estado e somando quantidade de exportações
    consulta_2021 = """
    SELECT importacoes.CO_ANO, uf.SG_UF, importacoes.CO_NCM, ncm.NO_NCM_POR,
        SUM(importacoes.QT_ESTAT) AS quantidade_total_importada
    FROM importacoes
    INNER JOIN ncm ON importacoes.CO_NCM = ncm.CO_NCM
    INNER JOIN uf ON importacoes.SG_UF_NCM = uf.SG_UF
    WHERE importacoes.CO_ANO = 2021
    GROUP BY uf.SG_UF, importacoes.CO_NCM
    ORDER BY uf.SG_UF, quantidade_total_importada DESC;
    """
    return consulta_2021
```

```python
conexao = sqlite3.connect('Balanco_comecial.db')
cursor = conexao.cursor()
exportacoes_2020 = pd.read_sql_query(consulta_exportacao(2020), conexao)
exportacoes_2021 = pd.read_sql_query(consulta_exportacao(2021), conexao)
importacoes_2020 = pd.read_sql_query(consulta_importacoes(2020), conexao)
importacoes_2021 = pd.read_sql_query(consulta_importacoes(2021), conexao)
conexao.close()
```

## Função responsavel por selecionar o top 3 em cada estado

```python
def selecionar_top3(dataframe, categoria):
    if categoria == 'exportacao':
        coluna_categoria = 'quantidade_total_exportada'
    else:
        coluna_categoria = 'quantidade_total_importada'

    # Ordenar o DataFrame pelo total exportado ou importado
    dataframe_ordenado = dataframe.sort_values(by=coluna_categoria, ascending=False)

    # Agrupar por estado e selecionar os top 3 produtos de cada estado
    top_3_por_estado = dataframe_ordenado.groupby('SG_UF').head(3)


    return top_3_por_estado
```

# Grafico top 3 por mes em cada estado

```python
# Conectar ao banco de dados SQLite
conexao = sqlite3.connect('Balanco_comecial.db')

# Consulta SQL para os top 3 produtos mais exportados por estado em 2021
consulta_2021 = """
SELECT exportacoes.CO_ANO, uf.SG_UF, exportacoes.CO_NCM, ncm.NO_NCM_POR,exportacoes.CO_MES,
       SUM(exportacoes.QT_ESTAT) AS quantidade_total_exportada
FROM exportacoes
INNER JOIN ncm ON exportacoes.CO_NCM = ncm.CO_NCM
INNER JOIN uf ON exportacoes.SG_UF_NCM = uf.SG_UF
WHERE exportacoes.CO_ANO = 2021
GROUP BY exportacoes.CO_ANO, uf.SG_UF,exportacoes.CO_MES, exportacoes.CO_NCM
ORDER BY uf.SG_UF, quantidade_total_exportada DESC;
"""

# Executar a consulta para 2021 e criar um DataFrame
exportacoes_2021_mes = pd.read_sql_query(consulta_2021, conexao)

# Fechar a conexão com o banco de dados
conexao.close()
```

```python
# # Pegar a parte da string que vem antes da primeira vírgula na coluna 'NO_NCM_POR'
exportacoes_2021_mes['NO_NCM_POR'] = exportacoes_2021_mes['NO_NCM_POR'].str.split(',').str[0].str.split().str[:3].str.join(' ')
```

```python
display(exportacoes_2021_mes[exportacoes_2021_mes['SG_UF']=='RN'])
```

```python
top3_por_mes_estado = exportacoes_2021_mes.groupby(['SG_UF', 'CO_MES']).apply(lambda x: x.nlargest(3, 'quantidade_total_exportada')).reset_index(drop=True)
```
